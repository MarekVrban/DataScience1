```python
df_age = df1.loc[(df1['Age']<100) & (df1['TotalWorkingYears']<100), ['Education', 'Gender','Age', 'TotalWorkingYears']].groupby(['Education','Gender']).agg('mean')
df_age.loc[:,'Start_age'] = df_age['Age'] - df_age['TotalWorkingYears']
df_age
```

```python
def get_age(ID, gender):
  education = float(df1.loc[df1['EmployeeNumber'] == ID, 'Education'])
  start_age = float(df_age.loc[education, gender]['Start_age'])
  TWrkYears = float(df1.loc[df1['EmployeeNumber'] == ID, 'TotalWorkingYears'])
  return int(start_age + TWrkYears)
```

```python
df1['Age'] = df1.apply(lambda x: x['Age'] if x['Age'] < 100 else get_age(x['EmployeeNumber'],x['Gender']),axis=1)
```

```python
df1['Age'].describe()
```

```python
def get_TWrkYears(ID, gender):
  education = float(df1.loc[df1['EmployeeNumber'] == ID, 'Education'])
  start_age = float(df_age.loc[education, gender]['Start_age'])
  Age = float(df1.loc[df1['EmployeeNumber'] == ID, 'Age'])
  years_at_company = float(df1.loc[df1['EmployeeNumber'] == ID, 'YearsAtCompany'])
  result = int(Age - start_age)
  return result if result > 0 else years_at_company
```

```python
df1['TotalWorkingYears'] = df1.apply(lambda x: x['TotalWorkingYears'] if x['TotalWorkingYears'] < 80 else get_TWrkYears(x['EmployeeNumber'],x['Gender']),axis=1)
```

```python
df1['TotalWorkingYears'].describe()
```

```python
def get_YrsComp(ID):   
  Age = float(df1.loc[df1['EmployeeNumber'] == ID, 'Age'])
  TWrkYears = float(df1.loc[df1['EmployeeNumber'] == ID, 'TotalWorkingYears'])
  NoCompWrk = float(df1.loc[df1['EmployeeNumber'] == ID, 'NumCompaniesWorked']) + 1
  promotion = int(df1.loc[df1['EmployeeNumber'] == ID, 'YearsSinceLastPromotion'])
  manager = int(df1.loc[df1['EmployeeNumber'] == ID, 'YearsWithCurrManager'])
  result = int(round(TWrkYears/NoCompWrk,0))
  return result if result > np.max([manager,promotion]) else np.max([manager,promotion])
```

```python
df1['YearsAtCompany'] = df1.apply(lambda x: x['YearsAtCompany'] if x['YearsAtCompany'] < 50 else get_YrsComp(x['EmployeeNumber']),axis=1)
```

```python
df1['YearsAtCompany'].describe()
```

```python
def get_YrsInRole(ID):  
  CompY = float(df1.loc[df1['EmployeeNumber'] == ID, 'YearsAtCompany'])  
  promotion = int(df1.loc[df1['EmployeeNumber'] == ID, 'YearsSinceLastPromotion'])
  manager = int(df1.loc[df1['EmployeeNumber'] == ID, 'YearsWithCurrManager'])
  if promotion ==0:
    if manager == 0:
      result = CompY
    else:
      result = manager
  else: 
    result = promotion
  
  return result
```

```python
df1['YearsInCurrentRole'] = df1.apply(lambda x: x['YearsInCurrentRole'] if x['YearsInCurrentRole'] < 50 else get_YrsInRole(x['EmployeeNumber']),axis=1)
```

```python
df1['YearsInCurrentRole'].describe()
```

```python
def des_object(df, varname):
    table = vcounts = df[varname].value_counts()
    vcounts_len = len(vcounts)    
    
    print("")
    print("*********************************")

    plt.figure(figsize=(16,7))
    plt.subplot(121)
    plt.axis([0, 10, 0, 10])
    i = 7.5
    plt.text(0, 8.5, f"there is: {vcounts_len} different values", horizontalalignment='left', fontsize=12 )
    for row in range(np.min([len(table),10])):
      text = table.index[row] + ": " + str(table.iloc[row])      
      plt.text(0, i, text, horizontalalignment='left', fontsize=12 )
      i-=0.5
   
    
    plt.text(0, 9, '-------------------------------------------------------', horizontalalignment='left', fontsize=12 )
    plt.title(varname + "| type:" + str(df[varname].dtype),loc='left', weight = 'bold')
    plt.axis('off')  
    plt.subplot(122)
    g = sns.countplot(x = varname, hue='Attrition', data = df)

    g.set_title('Attrition by category')
    plt.xlabel('')
    plt.xticks(rotation=90)      
    plt.tight_layout()
    plt.show()
    print("*********************************")
    print("")
    
def des_numeric(df, varname):
    print("*********************************")    
    table = pd.DataFrame(df[varname].describe().round(2))
    skw = skew(df[varname], axis=0, bias=True)
    kts = kurtosis(df[varname], axis=0, bias=True)

    plt.figure(figsize=(16,5))
    plt.subplot(131)
    plt.axis([0, 10, 0, 10])
    i = 8.5
    for row in range(len(table)):
      text = table.index[row] + ": " + str(table.iloc[row,0])      
      plt.text(0, i, text, horizontalalignment='left', fontsize=12 )
      i-=0.5
    
    
    plt.text(0, 9, '-------------------------------------------------------', horizontalalignment='left', fontsize=12 )
    plt.text(0, i, f"NA values: {df[varname].isna().sum()}", horizontalalignment='left', fontsize=12 )
    plt.text(0, i-0.5, f"unique values: {df[varname].nunique()}", horizontalalignment='left', fontsize=12 )
    plt.text(0, i-1.5, f"skew: {round(skw,2)}", horizontalalignment='left', fontsize=12 )
    plt.text(0, i-2, f"kurtosis: {round(kts,2)}", horizontalalignment='left', fontsize=12 )

    plt.title(varname + "| type:" + str(df[varname].dtype),loc='left', weight = 'bold')
    plt.axis('off')

    plt.subplot(132)
    g= sns.histplot(data=df[varname], kde = True)
    g.set_title('Histogram')
    plt.subplot(133)
    g1 = sns.boxplot(data=df[varname], palette=['#7FFF00'])
    g1.set_title('Boxplot')
    plt.xticks([])
    plt.tight_layout()
    plt.show()
    
def des_df(df):
    for c in df.columns:
        if df[c].dtype == object:
            des_object(df,c)
        else:
            des_numeric(df,c)
```

```python
from sklearn.model_selection import train_test_split

X = df1.drop('Attrition', axis=1)
y = df1.Attrition

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
X_train_ss = ss.fit_transform(X_train)
X_test_ss = ss.fit_transform(X_test)
```

```python
df1.info()
```

```python
from sklearn.linear_model import LogisticRegression

lg = LogisticRegression()
lg.fit(X_train_ss, y_train)
```

```python
pred = lg.predict(X_test_ss)
pred_proba = lg.predict_proba(X_test_ss)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, pred))
```

```python
from sklearn.ensemble import RandomForestClassifier


model = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=1)
model.fit(X_train_ss, y_train)
predictions = model.predict(X_test_ss)
predictions_proba = model.predict_proba(X_test_ss)
print(classification_report(y_test, predictions))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier
from xgboost.sklearn import XGBClassifier
#import xgboost as xgb

models = [
    LogisticRegression(),
    RandomForestClassifier(),
    SVC(probability=True),
    KNeighborsClassifier(),
    AdaBoostClassifier(),
    GradientBoostingClassifier(),
    XGBClassifier()
    ]
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import roc_auc_score
from sklearn.metrics import accuracy_score

#Train and score algorithms
for m in models:
    
    m.fit(X_train_ss, y_train)
    pred = m.predict(X_test_ss)
    auc = roc_auc_score(y_test, pred) 
    
    print(m.__class__.__name__)
    print("-----------------------")
    print(classification_report(y_test, pred))
    print('AUC-ROC score: ', auc)
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import plot_confusion_matrix
plot_confusion_matrix(model, X_test_ss, y_test)  
plt.show()
```
